```python
# 1 - Well Seperated Data
a_df = pd.read_csv("1_Well_separated_data\Well_Seprated_Class_A.csv")
a_df.drop("Width", inplace=True, axis=1)
b_df = pd.read_csv("1_Well_separated_data\Well_Seprated_Class_B.csv")
b_df.drop("Width", inplace=True, axis=1)
a_df['Label']='Class_A'
b_df['Label']='Class_B'
df_1 = pd.concat([a_df, b_df], ignore_index=True)
df_1['Type']='1_Well_separated_data'

# 2 - Well Seperated Data with Outliers
a_df = pd.read_csv("2_well_separated_data_wit_outliers\Well_Seprated_outliers_Class_A.csv")
a_df.drop("Width", inplace=True, axis=1)
b_df = pd.read_csv("2_well_separated_data_wit_outliers\Well_Seprated_outliers_Class_B.csv")
b_df.drop("Width", inplace=True, axis=1)
a_df['Label']='Class_A'
b_df['Label']='Class_B'
df_2 = pd.concat([a_df, b_df], ignore_index=True)
df_2['Type']='2_well_separated_data_wit_outliers'

# 3 - Concentric Circle Large Inner Circle
a_df = pd.read_csv("3_concentric_circles_large_inner\concentric_circles_large_inner_Class_A.csv")
a_df.drop("Width", inplace=True, axis=1)
b_df = pd.read_csv("3_concentric_circles_large_inner\concentric_circles_large_inner_Class_B.csv")
b_df.drop("Width", inplace=True, axis=1)
a_df['Label']='Class_A'
b_df['Label']='Class_B'
df_3 = pd.concat([a_df, b_df], ignore_index=True)
df_3['Type']='3_concentric_circles_large_inner'

# 4 - Concentric Circle Small Inner Circle
a_df = pd.read_csv("4_concentric_circles_small_inner\concentric_circles_small_inner_Class_A.csv")
a_df.drop("Width", inplace=True, axis=1)
b_df = pd.read_csv("4_concentric_circles_small_inner\concentric_circles_small_inner_Class_B.csv")
b_df.drop("Width", inplace=True, axis=1)
a_df['Label']='Class_A'
b_df['Label']='Class_B'
df_4 = pd.concat([a_df, b_df], ignore_index=True)
df_4['Type']='4_concentric_circles_small_inner'

# 5 - S Curve
a_df = pd.read_csv("5_S_Curve\S_curve_class_A.csv")
a_df.drop("Width", inplace=True, axis=1)
b_df = pd.read_csv("5_S_Curve\S_curve_class_B.csv")
b_df.drop("Width", inplace=True, axis=1)
a_df['Label']='Class_A'
b_df['Label']='Class_B'
df_5 = pd.concat([a_df, b_df], ignore_index=True)
df_5['Type']='5_S_Curve'

# 6 - EX_OR
a_df = pd.read_csv("6_EX_OR\EX_OR_Class_A.csv")
a_df.drop("Width", inplace=True, axis=1)
b_df = pd.read_csv("6_EX_OR\EX_OR_Class_B.csv")
b_df.drop("Width", inplace=True, axis=1)
a_df['Label']='Class_A'
b_df['Label']='Class_B'
df_6 = pd.concat([a_df, b_df], ignore_index=True)
df_6['Type']='6_EX_OR'

# 7 - Spiral
a_df = pd.read_csv("7_Spiral\Spiral_Class_A.csv")
a_df.drop("Width", inplace=True, axis=1)
b_df = pd.read_csv("7_Spiral\Spiral_Class_B.csv")
b_df.drop("Width", inplace=True, axis=1)
a_df['Label']='Class_A'
b_df['Label']='Class_B'
df_7 = pd.concat([a_df, b_df], ignore_index=True)
df_7['Type']='7_Spiral'
```

```python
df = pd.concat([df_1, df_2, df_3, df_4, df_5, df_6, df_7], ignore_index=True)
g = sns.FacetGrid(df, hue="Label", col="Type", col_wrap=3, size=7, margin_titles=True)
g.map(plt.scatter, "X", "Y", alpha=.7)
g.add_legend()
plt.title("1_Well_separated_data")
plt.show()
```
