### Q-Q Plots

In statistics, a Q–Q plot is a probability plot, which is a graphical method for comparing two probability distributions by plotting their quantiles against each other.     
It helps us in understanding whether two random variables belong from same distribution.                         
If the two distributions being compared are similar, the points in the Q–Q plot will approximately lie on the line y = x.    
If the distributions are linearly related, the points in the Q–Q plot will approximately lie on a line, but not necessarily on the line y = x.                   

The points plotted in a Q–Q plot are always non-decreasing when viewed from left to right. If the two distributions being compared are identical, the Q–Q plot follows the 45° line y = x. If the two distributions agree after linearly transforming the values in one of the distributions, then the Q–Q plot follows some line, but not necessarily the line y = x. If the general trend of the Q–Q plot is flatter than the line y = x, the distribution plotted on the horizontal axis is more dispersed than the distribution plotted on the vertical axis.

The purpose of the quantile-quantile (QQ) plot is to show if two data sets come from the same distribution.
Plotting the first data set’s quantiles along the x-axis and plotting the second data set’s quantiles along the
y-axis is how the plot is constructed. In practice, many data sets are compared to the normal distribution. The
normal distribution is the base distribution and its quantiles are plotted along the x-axis as the “Theoretical
Quantiles” while the sample quantiles are plotted along the y-axis as the “Sample Quantiles”

```python
import numpy as np 
import pylab 
import scipy.stats as stats
from IPython.display import Latex

measurements = np.random.normal(loc = 20, scale = 5, size=100)   
stats.probplot(measurements, dist="norm", plot=pylab)
pylab.show()
```

```python
import statsmodels.api as sm
test = np.random.normal(0,1, 1000)

sm.qqplot(test, line='45')
pylab.show()
```

Q-Q Plots are non-parametric whereas Normal plots as well as Cumulative distribution function are parametric. Hence plotting cdfs of random variable x and y will result in similar plots if the variables belong from similar distribution. 

### Example:

Let us consider the following:                                                                                      
X=heights of all men in Hyderabad(mu=150,sigma=10)                                                                      
Y=heights of all men in Delhi(mu=152,sigma=10)

```python
#plotting pdfs in for the given datas
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
mu1 = 150
mu2 = 152
sigma = 10
x = np.linspace(mu1-40,mu1+40, 100)
plt.plot(x,mlab.normpdf(x, mu1, sigma))
plt.plot(x,mlab.normpdf(x, mu2, sigma))

plt.show()
```

```python
#plotting Q-Q plots for the given data:

import statsmodels.graphics.gofplots as gof

test1 = np.random.normal(150, 10, 100)
test2 = np.random.normal(152, 10, 100)
quantile1 = test1
quantile2 = test2
gof.qqplot_2samples(quantile1 , quantile2, xlabel="x samples", ylabel="y samples")
pylab.show()
```

#### Application of Normalization/Standarization

The Q-Q Plot obtained above can be normalized to X' and Y' by applying centering and scaling and then the plot we will obtain will be a 45 degree line where x=y.

### Log-Normal Distribution

In probability theory, a log-normal (or lognormal) distribution is a continuous probability distribution of a random variable whose logarithm is normally distributed. Thus, if the random variable X is log-normally distributed, then Y=ln(X) has a normal distribution. Likewise, if Y has a normal distribution, then the exponential function of Y,X=exp(Y), has a log-normal distribution. A random variable which is log-normally distributed takes only positive real values. The distribution is occasionally referred to as the Galton distribution or Galton's distribution, after Francis Galton.




## Covariance

The covariance between two jointly distributed real-valued random variables X and Y with finite second moments is defined as the expected product of their deviations from their individual expected values.

 cov(X,Y)=E[(X-E[X])(Y-E[Y])],
 where E[X] is the expected value of X, also known as the mean of X.
 

```python
Latex(r"""\begin{eqnarray}
\operatorname {cov} (X,Y)=\operatorname {E} {{\big [}(X-\operatorname {E} [X])(Y-\operatorname {E} [Y]){\big ]}}
\end{eqnarray}""")
```

```python
Latex(r"""\begin{eqnarray}
{\displaystyle {\begin{aligned}\operatorname {cov} (X,Y)&=\operatorname {E} \left[\left(X-\operatorname {E} \left[X\right]\right)\left(Y-\operatorname {E} \left[Y\right]\right)\right]\\&=\operatorname {E} \left[XY-X\operatorname {E} \left[Y\right]-\operatorname {E} \left[X\right]Y+\operatorname {E} \left[X\right]\operatorname {E} \left[Y\right]\right]\\&=\operatorname {E} \left[XY\right]-\operatorname {E} \left[X\right]\operatorname {E} \left[Y\right]-\operatorname {E} \left[X\right]\operatorname {E} \left[Y\right]+\operatorname {E} \left[X\right]\operatorname {E} \left[Y\right]\\&=\operatorname {E} \left[XY\right]-\operatorname {E} \left[X\right]\operatorname {E} \left[Y\right].\end{aligned}}}
\end{eqnarray}""")
```

## Pearson correlation coefficient

Pearson's correlation coefficient is the covariance of the two variables divided by the product of their standard deviations. The form of the definition involves a "product moment", that is, the mean (the first moment about the origin) of the product of the mean-adjusted random variables; hence the modifier product-moment in the name.

rho{X,Y}=E[(X-mu_X)(Y-\mu_Y)]/sigma _X*sigma _Y

   where mu_X is the mean of X,
           E= expectation
           


```python
Latex(r"""\begin{eqnarray}
\rho _{X,Y}={\frac {\operatorname {cov} (X,Y)}{\sigma _{X}\sigma _{Y}}}
\end{eqnarray}""")
```



PCC tells us how two variables are correlated.                                         
+1 implies perfectly correlated                                                                           
-1 implies pefectly non correlated

### Limitations:

PCC is only good for linear relationship. It fails to give a proper measurement for non-linear relationships.
For this reason, when there is outliers in the data, PCC is completely useless.

## Spearman's rank correlation coefficient


The Spearman correlation coefficient is defined as the Pearson correlation coefficient between the ranked variables;while Pearson's correlation assesses linear relationships, Spearman's correlation assesses monotonic relationships.
 If there are no repeated data values, a perfect Spearman correlation of +1 or −1 occurs when each of the variables is a perfect monotone function of the other.

```python

Latex(r"""\begin{eqnarray}
{\displaystyle r_{s}=\rho _{\operatorname {rg} _{X},\operatorname {rg} _{Y}}={\frac {\operatorname {cov} (\operatorname {rg} _{X},\operatorname {rg} _{Y})}{\sigma _{\operatorname {rg} _{X}}\sigma _{\operatorname {rg} _{Y}}}}}
\end{eqnarray}""")
```

## Correlation vs Causation

#### Correlation: 
Correlation is any of a broad class of statistical relationships involving dependence, though in common usage it most often refers to the extent to which two variables have a linear relationship with each other.

Example: orrelation between the physical statures of parents and their offspring, and the correlation between the demand for a product and its price.

#### Causation:
Causation refers to the fact that whether one group of data is responsible for the behaviour of another group of data.

Example: The Plasmodium parasite that causes malaria.

#### If two random variables have correlation,it doesn't mean that one variable causes another.
